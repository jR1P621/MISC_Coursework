```python
import pandas as pd

raw_data = pd.read_csv('Skyserver_SQL2_27_2018 6_51_39 PM.csv')
raw_data.head()
```

```python
y = raw_data.pop('class').values
X = raw_data.values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# # Train and find fits

# tree_scores = pd.DataFrame()

# for test_percent in range(10, 100, 10):
#     percent_results = []
#     for tree_depth in range(1, 18):
#         depth_runs = []
#         for i in range(0, 10):
            
#             X_train, X_test, y_train, y_test = train_test_split(
#                 X, y, test_size=(test_percent/100), stratify=y)
#             dec_tree = DecisionTreeClassifier(max_depth=tree_depth)
#             dec_tree.fit(X_train, y_train)
#             depth_runs.append(dec_tree.score(X_test, y_test))
                        
#         ave_score = sum(depth_runs)/(i+1)
#         percent_results.append(ave_score)
        
#     tree_scores[test_percent] = percent_results
    
# tree_scores.to_csv('tree_scores.csv', index=False,)
```

```python
# Create Tree

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y)

dec_tree = DecisionTreeClassifier(max_depth=2, max_leaf_nodes=3)
dec_tree.fit(X_train, y_train)
```

```python
# Visualize Tree

from sklearn.tree import export_graphviz
from graphviz import Source
from numpy import unique

tree_img = Source(export_graphviz(
    dec_tree,
    out_file=None,
    feature_names=raw_data.columns,
    class_names=sorted(sorted(unique(y))),
    rounded=True,
    filled=True))
```
