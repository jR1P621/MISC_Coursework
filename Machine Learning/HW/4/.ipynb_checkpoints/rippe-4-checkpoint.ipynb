```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.metrics import categorical_crossentropy

import tensorflow_datasets as tfds

import seaborn as sns

from sklearn.metrics import confusion_matrix
```

```python
tf.config.list_physical_devices()
```

```python
img_height = 150
img_width = 150
data_dir = 'seg_'
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir + 'train',
    image_size=(img_height, img_width)
)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir + 'test',
    image_size=(img_height, img_width)
)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
for image_batch, labels_batch in val_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
image_batch, labels_batch = next(iter(train_ds))
first_image = image_batch[0]
print(np.min(first_image), np.max(first_image))
```

```python
train_labels = np.concatenate([y for x, y in train_ds], axis=0)
val_labels = np.concatenate([y for x, y in val_ds], axis=0)
```

```python
#Check our class distribution
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
classes, counts = np.unique(train_labels, return_counts=True)
plt.barh(class_names, counts)
plt.title('Class distribution in training set')

plt.subplot(2, 2, 2)
classes, counts = np.unique(val_labels, return_counts=True)
plt.barh(class_names, counts)
plt.title('Class distribution in validation set')
```

```python
train_ds_16 = train_ds.unbatch().batch(16).cache().prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds_16 = val_ds.unbatch().batch(16).cache().prefetch(buffer_size=tf.data.AUTOTUNE)
train_ds_32 = train_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds_32 = val_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
train_ds_64 = train_ds.unbatch().batch(64).cache().prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds_64 = val_ds.unbatch().batch(64).cache().prefetch(buffer_size=tf.data.AUTOTUNE)

num_classes = 6
```

```python
# Test model to see model shapes
test_model = tf.keras.Sequential([
    layers.experimental.preprocessing.Rescaling(1./255, input_shape=(150,150,3)),
    layers.Conv2D(32, 3, activation='sigmoid'),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(num_classes, activation='softmax')
])
test_model.summary()
```

```python
models = {}
results = {}
```

```python
models['sigmoid'] = tf.keras.Sequential([
    layers.experimental.preprocessing.Rescaling(1./255, input_shape=(150,150,3)),
    layers.Conv2D(32, 3, activation='sigmoid'),
    layers.MaxPool2D(),
    layers.Flatten(),
    layers.Dense(num_classes, activation='softmax')
])
models['sigmoid'].compile(
    optimizer='adam',
    loss=tf.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
results['sigmoid'] = models['sigmoid'].fit(
    train_ds_32,
    validation_data=val_ds_32,
    epochs=15
)
```

```python
for h in results:
    acc = results[h].history['accuracy']
    val_acc = results[h].history['val_accuracy']
    loss = results[h].history['loss']
    val_loss = results[h].history['val_loss']
    epochs_range = range(15)

    plt.figure(figsize=(15, 15))
    ax = plt.subplot(2, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    ax.set_xlabel('epoch')
    ax.set_ylabel('accuracy')
    ax.set_ybound(upper=1,lower=.5)
    plt.title('Training and Validation Accuracy (' + h + ')')

    ax = plt.subplot(2, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    ax.set_xlabel('epoch')
    ax.set_ylabel('loss')
    plt.title('Training and Validation Loss (' + h + ')')
    plt.show()
```

```python
import os

classes = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']
classes_label = {classes:i for i, classes in enumerate(classes)}

path = data_dir + 'test'
val_images = []
val_labels = []

# Iterate through each folder corresponding to a category
for folder in os.listdir(path):
    if os.path.isdir(os.path.join(path, folder)):
        label = classes_label[folder]

        # Iterate through each image in our folder
        for file in os.listdir(os.path.join(path, folder)):
            if '.jpg' in file:
                # Get the path name of the image
                img_path = os.path.join(os.path.join(path, folder), file)
                #print(img_path)

                # Open the img
                image = tf.keras.preprocessing.image.load_img(path=img_path, target_size=(img_width, img_height))
                input_arr = tf.keras.preprocessing.image.img_to_array(image)
                input_arr = np.array([input_arr])

                # Append the image and its corresponding label to the output
                val_images.append(input_arr)
                val_labels.append(label)
```

```python
val_pred = []
for i in range(0,len(val_images)):
    val_pred.append(np.argmax(models['exp9'](val_images[i])))
```

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
sns.heatmap(confusion_matrix(val_labels, val_pred), annot=True,
            annot_kws={"size": 15},
            xticklabels=classes, 
           yticklabels=classes, ax = ax,
           fmt='g')
ax.set_title('Confusion matrix')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
plt.show()
```

```python
C = confusion_matrix(val_labels, val_pred)
C = 100*C / C.astype(np.float).sum(axis=0)
plt.figure(figsize=(10,7))
ax = plt.axes()
sns.heatmap(C, annot=True,
            annot_kws={"size": 15},
            xticklabels=classes, 
           yticklabels=classes, ax = ax,
           fmt='.0f')
ax.set_title('Confusion matrix (%)')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.metrics import categorical_crossentropy

import tensorflow_datasets as tfds

from os import walk
from random import randint


loaded_model = tf.keras.models.load_model('saved_model')
classes = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']

path = data_dir + 'pred/'
_, _, filenames = next(walk(path))
```

### Run me to generate a test of random files

```python
import matplotlib.gridspec as gridspec

fig_size = 2
fig = plt.figure(figsize=(30, 30))
outer = gridspec.GridSpec(fig_size, fig_size*2, wspace=0.2, hspace=0.2)

for i in range(fig_size*fig_size*2):
    inner = gridspec.GridSpecFromSubplotSpec(2, 1,subplot_spec=outer[i], wspace=0.1, hspace=0.1)
    file = randint(0, len(filenames))
    path = data_dir + 'pred/' + filenames[file]
    image = tf.keras.preprocessing.image.load_img(path=path, target_size=(img_width, img_height))
    input_arr = tf.keras.preprocessing.image.img_to_array(image)
    input_arr = np.array([input_arr])
    y_pred = loaded_model.predict(input_arr)
    pred_class = classes[np.argmax(y_pred)]
    pred_prob = y_pred.reshape(6)
    for j in range(2):
        if (j%2) == 0:
            ax = plt.Subplot(fig, inner[j])
            ax.imshow(image)
            ax.set_title(pred_class)
            ax.set_xticks([])
            ax.set_yticks([])
            fig.add_subplot(ax)
        else:
            ax = plt.Subplot(fig, inner[j])
            ax.bar(classes,pred_prob)
            fig.add_subplot(ax)
            ax.set_ylabel('probability')
```
